## A Mini Project on IPL Data Set


### 1. To display the number of matches yearwise.

```python
import pandas as pd
filepath ='Match.csv'
def readCsvdata(filepath):
    return pd.read_csv(filepath)
df=readCsvdata(filepath)
print(df) 
```

### 2. Plot a bar graph  on no. of matches won by each team in the year 2011

```python
import matplotlib.pyplot as plt 
pt=df[df['Season_Year']==2011]
lr=pt['match_winner'].value_counts()
pd.DataFrame(lr)
print(lr)
print()
teams=[]
teams=lr.index.tolist()
plt.barh(teams,lr,color='blue')
```

### 3.Print  ManOfMatch count of each player in  Hyderabad stadium

```python
df2=df.copy()
adf=pd.DataFrame(df2[df2['City_Name'].isin(['Hyderabad'])])
pd.DataFrame(adf.groupby('ManOfMach')['Match_SK'].count())
```
